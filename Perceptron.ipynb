**A line through the origin**

```
state = 0

M = [2, -4]

# equation of a line
def get_y_throughorigin(x1) : 
   return (x1*M[0]) / -M[1]

# create a list of the numbers between -10 and 10
x1List = list(range(-10,10))
```

```
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
plt.axis('equal')

state += 1

# create a list with the results of the equation of a line for these numbers
x2oList = []
for x in x1List:
  y1 = get_y_throughorigin(x)  
  x2oList.append(y1)


if state > 0:
  # plotting the weight vector as a line
  plt.plot([0, M[0]], [0, M[1]], 'k-')

if state > 1:
  
  # plotting the results of the equation of a line as black dots
  plt.plot(x1List[:min(state-1,20)], x2oList[:min(state-1,20)], 'o', color='green');
  plt.arrow(0, 0, x1List[min(state-2,20)], x2oList[min(state-2,20)], head_width=0.3, head_length=0.6, fc='r', ec='r', length_includes_head = True);
  
```

**A line at a distance from the origin**

```
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
plt.axis('equal')

state += 1

# create a list with the results of the equation of a line for these numbers
x2dList = []
for x in x1List:
  y1 = get_y_atdistance(x)  
  x2dList.append(y1)


if state > 0:
  # plotting the weight vector as a line
  plt.plot([0, M[0]], [0, M[1]], 'k-')

if state > 1:
  
  # plotting the results of the equation of a line as black dots
  plt.plot(x1List[:min(state-1,20)], x2dList[:min(state-1,20)], 'o', color='green');
  plt.arrow(0, 0, x1List[min(state-2,20)], x2dList[min(state-2,20)], head_width=0.3, head_length=0.6, fc='r', ec='r', length_includes_head = True);
```

**Perceptron equation**

```
import numpy as np
from functools import reduce

# Create a list of values going from -5 to +5 (that are 11 values)
cx = list(range(-5,6))

def list_from_value(v):
  # create a list of 11 times the same value
  vlist = [v] * 11
  #print(vlist)
  
  # Create a list of values going from -5 to +5
  llist = list(range(-5,6))
  #print(llist)
  #return [list(zip(vlist, llist))]
  
  # Create a list of tuples combinging each element of vlist with an element of llist
  # zip in python > https://medium.com/@happymishra66/zip-in-python-48cb4f70d013
  return [list(a) for a in zip(vlist, llist)]

# map: apply a function to all items in the input
# list: create a list of values
cres = list(map(list_from_value, cx))

combined_list = reduce(lambda x,y: x+y,cres)

x1 = np.array(combined_list)[:, 0]
x2 = np.array(combined_list)[:, 1]
```

```

def get_preceptron_dotproduct(x1, x2, wght) :
  return x1 * wght[0] + x2 * wght[1] + wght[2]

def get_preceptron_result(x1, x2, wght) :
  dotproduct_result = get_preceptron_dotproduct(x1, x2, wght)
  return 1 if dotproduct_result > 0 else 0
```

```
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
plt.style.use('seaborn-whitegrid')

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

w_perceptron = [6, -12, 2]
perceptron_result = []

for c in combined_list:
  r = get_preceptron_result(c[0], c[1], w_perceptron)  
  perceptron_result.append(r)
  
ax.scatter(x1, x2, perceptron_result, c='b', marker='o', label='perceptron result')

ax.set_xlabel('x1 axis')
ax.set_ylabel('x2 axis')
ax.set_zlabel('perceptron outcome')
plt.title("Perceptron")
plt.legend()
plt.tight_layout()
plt.show()
```

```
def get_perceptron_x2(x1, w) :
  return (x1*w[0] + w[2]) / -w[1]

def perceptron_learning(x1, x2, x3, wn, wd, learning_rate) :
  desired_outcome = get_preceptron_result(x1, x2, wd)
  #print(desired_outcome)
  current_outcome = get_preceptron_result(x1, x2, wn)  
  #print(current_outcome)

  data_point = [x1, x2, x3];
  if current_outcome != desired_outcome:
    new_learning = wn +  [i * (desired_outcome - current_outcome) * learning_rate for i in data_point]
    #print("new_learning" + str(new_learning))
    return new_learning
  
  #print("no learning" + str(w))
  return wn
```

**Perceptron learning: case 1**

```
import numpy as np

state = 0

learning_rate = 1
wn = np.array([-1, 6, 0])
wd = np.array([-1, 1, 0])
```

```
feature = np.array([3, 1, 0])
```

```
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
plt.axis('equal')
#plt.axis([-10, 10, -10, 10])
plt.xticks(np.arange(-10, 11, step=1))
plt.yticks(np.arange(-10, 11, step=1))
state += 1

plt.plot([-5, 5], [get_perceptron_x2(-5, wn), get_perceptron_x2(5, wn)], 'r:')
plt.plot([-5, 5], [get_perceptron_x2(-5, wd), get_perceptron_x2(5, wd)], 'g-')

if state > 0:
  plt.arrow(0, 0, wd[0], wd[1], head_width=0.3, head_length=0.6, fc='g', ec='g', length_includes_head = True)
  plt.arrow(0, 0, wn[0], wn[1], head_width=0.3, head_length=0.6, fc='r', ec='r', length_includes_head = True)
  
if state > 1:
  plt.arrow(0, 0, feature[0], feature[1], head_width=0.3, head_length=0.6, fc='b', ec='b', length_includes_head = True)
  
if state > 2:
  plt.arrow(wn[0], wn[1], -1 * feature[0], -1 * feature[1], head_width=0.3, head_length=0.6, fc='b', ec='b', length_includes_head = True)
  
if state > 3:
  wn = perceptron_learning(feature[0], feature[1], feature[2], wn, wd, 1)
  plt.arrow(0, 0, wn[0], wn[1], head_width=0.3, head_length=0.6, fc='b', ec='b', length_includes_head = True)
  state = 0
```
